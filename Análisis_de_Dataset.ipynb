# **Análisis de Dataset**

```python
#Cargamos las librerias
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
#Cargamos el DataFrame utilizado anteriormente en la tarea 3
hypertension=pd.read_csv('https://raw.githubusercontent.com/Andy-ArandaL/TSC/main/healthcare-dataset-stroke-data.csv')
```

```python
#Se muestran los datos iniciales de nuestro dataframe
hypertension.head()
```

```python
#Definimos otro DataFrame
hypertension1=hypertension
```

```python
#Empezamos a reemplazar los datos de tipo caracter a tipo numerico
df=hypertension1.replace({"Yes": 1, "No": 0})
df
```

```python
#Vemos cuales son los tipos de caracteres que exiten en la columna de Genero
hypertension1['gender'].unique()
```

```python
#Reemplazamos los caracteres de genero por numeros, donde 0=Mujer, 1=Hombre y 2=Otro
df=df.replace({"Male": 1, "Female": 0,"Other":2})
df
```

```python
#Vemos los tipos de datos que contiene la columna de "smoking_status"
hypertension1['smoking_status'].unique()
```

```python
#Reemplazamos los datos de smoking por numericos
df=df.replace({"formerly smoked":1,"never smoked":0,"smokes":2,"Unknown":3})
df
```

```python
#Vemos cuales son los tipos de caracteres que exiten en la columna de Work_type
hypertension1['work_type'].unique()
```

```python
#Reempazamos los datos de tipo de empleo por variables numericas
df=df.replace({"Private": 1, "Self-employed": 2, "Govt_job": 3, "children":4, "Never_worked":5})
df
```

```python
#Eliminamos la columna de Residence_type, ya que consideramos que no era necesario agregarla
df=df.drop('Residence_type', axis=1)
```

```python
df
```

```python
df=df.drop('id', axis=1)
```

```python
df
```

```python
#Eliminamos los valores nulos, ya que si los dejamos no nos dejara continuar con el procedimiento
df=df.dropna()
```

```python
#Reiniciamos el indice, por lo datos que se eliminaron que eran nulos
df=df.reset_index()
```

```python
df
```

```python
#Eliminamos la columna "index" ya que no nos sirve
df=df.drop('index', axis=1)
```

```python
df
```

```python
#Vemos cuales valores existen en la columna de "Stroke" donde 1=Si y 0=No
df['stroke'].unique()
```

```python
#Cuantas veces se repoten los datos de la columna de "Stroke"
df['stroke'].value_counts()
```

```python
#Definimos a "X" al eliminar la columna de "Stroke" y nos muestra los primeros valores del DataFrame
X=df.drop('stroke',axis=1)
X.head()
```

```python
#Hacemos Y quitando las columnas que no necesitamos de df
Y=df.drop(['gender', 'age', 'hypertension','heart_disease', 'ever_married','work_type','avg_glucose_level', 'bmi', 'smoking_status'], axis=1)
```

```python
#Volvemos los valores de Y de enteros a objetos
Y=Y.replace({1:"Yes", 0:"No"})
Y
```

```python
Y['stroke'].value_counts()
```

```python
#Importamos la siguiente libreria que nos sirve para dividir nuestro DataFrame en bloques (tipo matriz)
from sklearn.model_selection import train_test_split
```

```python
#seccionamos nuestro DataFrame
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.65)
```

```python
#como pasamos como argumento a train_test_split el dataset completo, des devuelve dos subconjuntos
#del mismo, escogiendo un cierto numero de filas para el subconjunto de train (entrenamiento) y
#el resto para el subconjunto de test (pruebas)
X_train
```

```python
X_test
```

```python
#Vemos el tamaño de nuestro Dataset
X.shape
```

```python
#la siguiente libreria la usamos para atributos categoricos, este criterio intenta estimar la informacion que aporta cada atributo basado
#en la informacion dada, tambien para medir la aleatoriedad de incertidumbre de un valor aleatorio de una variable "x"
from sklearn.tree import DecisionTreeClassifier
```

```python
#definimos arbol pues buscamos un arbol de clasificacion
arbol=DecisionTreeClassifier(max_depth=5)
```

```python
arbol=arbol.fit(X_train,Y_train)
```

```python
#definimos a "predicciones" como el arbol de prediccion para las pruebas de x
predicciones=arbol.predict(X_test)
```

```python
predicciones
```

```python
#Mandamos llamar los primeros datos para "Y-prueba"
Y_test.head()
```

```python
Y_test['stroke'].value_counts()
```

```python
#Vemos el tamaño del dataset de Y-prueba
Y_test.shape
```

```python
#Para construir un infome de texto que muestre las principales metricas.
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(Y_test, predicciones))
```

```python
#Generamos una matriz donde cada columa de la matriz es numero de predicciones de cada clase
print(confusion_matrix(Y_test,predicciones))
```

```python
#construimos una grafica para ver el comportamiento de los datos dada la matriz
sns.heatmap(confusion_matrix(Y_test,predicciones),annot=True)
```

```python
#Para generar metricas
from sklearn import metrics
```

```python
#La metrica "Accuracy" representa el porcentaje total de los valores correctamente
#clasificados
print("Accuracy:",metrics.accuracy_score(Y_test, predicciones))
```

```python
#Finalmente para generar el arbol a partir de lo ya obtenido
from sklearn import tree
```

```python
#Hacemos esta lista para ponerla en el diagrama de arbol y darle formato
lst=list(X.columns.values)
lst
```

```python
#Esta tambien es para darle formato al diagrama
lst1=['No','Yes']
```

```python
import matplotlib.pyplot as plt

f=plt.figure(figsize=(60,23))
tree.plot_tree(arbol, feature_names=lst, class_names=lst1, filled=True, fontsize=12)
plt.show()
```
